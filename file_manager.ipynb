```python
file_name = input('enter the file name: ')
qty = int(input('enter the no of files you want: '))
file_format = input('enter the file format: ')
```

```python
files = []
```

```python
for num in range(1,qty+1):
    file = open(f'{file_name}{num}{file_format}', 'x')
    files.append(f'{file_name}{num}{file_format}')
    
```

```python
import os
```

```python
files
```

```python
for file in files:
    os.remove(file)
```
